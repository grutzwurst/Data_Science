```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

**Cups of coffee per day** ;)<br>
0 = Monday<br>
.<br>
.<br>
.<br>
6 = Sunday

```python
weekday = np.arange(0, 7)
cups_coffee = np.array([5, 5, 4, 3, 4, 3, 3])

plt.scatter(weekday, cups_coffee)
```

Equation to solve:<br>
cups_coffee = m\*weekday + c<br>
We need to find *m* (slope) and *c* (offset):

```python
A = np.vstack([weekday, np.ones(len(weekday))]).T
m, c = np.linalg.lstsq(A, cups_coffee, rcond=None)[0]
```

```python
m, c
```

```python
plt.scatter(weekday, cups_coffee)
plt.plot(weekday, m*weekday + c, 'r')
plt.show()
```
